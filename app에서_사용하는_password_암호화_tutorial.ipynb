# 암호환된 Password 저장

#### 아무 환경에서 암호화된 Password 저장

```python
with open("password.txt", "r") as f:
    client_info = f.read()
    
cryp_password = cryptocode.encrypt(client_info, "사용할 마스터 키")

print(cryp_password)

with open("cryped_password", "w") as f:
    f.write(cryp_password)
```

# 사용할 APP에서 암호화된 Password 복호화

```python
master_key = os.environ['master_key']
```

```python
with open("cryped_password", "r") as f:
    cryp_password = f.read()
```
